```python
import pandas as pd
import numpy as np
```

```python
# load data
path_tweets = "data/tweets/"
path_graphs = "data/graphs/"

# tweets
tweets1 = pd.read_csv(path_tweets+"anthrocon_tweets.csv", delimiter=',')
tweets2 = pd.read_csv(path_tweets+"comiccon2017_tweets.csv", delimiter=',')
tweets3 = pd.read_csv(path_tweets+"icann2016_tweets.csv", delimiter=',')

# edges
edges1 = np.genfromtxt(path_graphs+"anthrocon_edgelist.csv", dtype=int, delimiter=',')
edges2 = np.genfromtxt(path_graphs+"comiccon2017_edgelist.csv", dtype=int, delimiter=',')
edges3 = np.genfromtxt(path_graphs+"icann2016_edgelist.csv", dtype=int, delimiter=',')
```

```python
# tweets

print("tweets1:")
display(tweets1.describe())
print("users: ", tweets1.node_id.unique().size)
print("tweets: ", tweets1.tweet_id.unique().size)

print("\ntweets2:")
display(tweets2.describe())
print("users: ", tweets2.node_id.unique().size)
print("tweets: ", tweets2.tweet_id.unique().size)

print("\ntweets3:")
display(tweets3.describe())
print("users: ", tweets3.node_id.unique().size)
print("tweets: ", tweets3.tweet_id.unique().size)
```
