```python
import pandas as pd
import numpy as np
```

## generate labels from tweet data

### binary labels:
* 0: did not attend anthrocon
* 1: did attended anthrocon

```python
# load data
path = "data/tweets/"

tweets1 = pd.read_csv(path+"anthrocon_tweets.csv", delimiter=',')
tweets2 = pd.read_csv(path+"comiccon2017_tweets.csv", delimiter=',')
tweets3 = pd.read_csv(path+"icann2016_tweets.csv", delimiter=',')
```

```python
def get_nodes(tweets):
    return tweets[["node_id", "attendance_label"]].groupby("node_id").max().reset_index()
```

```python
nodes1 = get_nodes(tweets1)
nodes2 = get_nodes(tweets2)
nodes3 = get_nodes(tweets3)
```

```python
print("nodes1: ")
display(nodes1.describe())
display(nodes1.head())

print("nodes2: ")
display(nodes2.describe())
display(nodes2.head())

print("nodes3: ")
display(nodes3.describe())
display(nodes3.head())
```

```python
# save labels
path = "data/nodes/"

np.savetxt(path+'twitter_anthrocon.labels', nodes1, delimiter='\t', fmt=['%i', '%i'])
np.savetxt(path+'twitter_comiccon2017.labels', nodes2, delimiter='\t', fmt=['%i', '%i'])
np.savetxt(path+'twitter_icann2016.labels', nodes3, delimiter='\t', fmt=['%i', '%i'])
```
