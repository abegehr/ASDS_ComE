```python
import pandas as pd
import numpy as np
```

## generate labels from tweet data

### binary labels:
* 0: did not attend anthrocon
* 1: did attended anthrocon

```python
# load data
path = "data/tweets/"

anthrocon_tweets = pd.read_csv(path+"anthrocon_tweets.csv", delimiter=',')
#tweets2 = pd.read_csv(path+"comiccon2017_tweets.csv", delimiter=',')
#tweets3 = pd.read_csv(path+"icann2016_tweets.csv", delimiter=',')
```

```python
display(anthrocon_tweets.describe())
display(anthrocon_tweets.head())
```

```python
nodes = anthrocon_tweets[["node_id", "attendance_label"]].groupby("node_id").max().reset_index()
```

```python
display(nodes.describe())
display(nodes.head())
```

```python
# save anthrocon labels
np.savetxt('twitter_anthrocon.labels', nodes, delimiter='\t', fmt=['%i', '%i'])
```
